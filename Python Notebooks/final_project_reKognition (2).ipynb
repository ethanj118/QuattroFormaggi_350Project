We chose four emotions for tests -- Confused, Disgusted, Calm and Surprise. We first selected the pictures with matched emotions. For each picture, we examined the confidence level for its matched emotions and also listed the most possible emotion detected by Rekognition.

```python
# test for confused emotion
confused = ['IMG_1045.jpg','IMG_1046.jpg','IMG_1056.jpg'] #select the pictures with confused emotion
detect = {}
confidence_confused = []
for name in confused:
    result = client.detect_faces(Image = {'S3Object':{'Bucket':'rekognition-project-files','Name':name}}, Attributes = ["ALL"])
    emotions = result['FaceDetails'][0]['Emotions']
    if emotions[0]['Type'] in detect:
        detect[emotions[0]["Type"]] += 1
    else:
        detect[emotions[0]["Type"]] = 1
    for emotions_type in emotions:
        if emotions_type["Type"] == "CONFUSED":
            confidence_confused.append(emotions_type['Confidence'])

print("Computer detected:")
print(detect)
print("Confidence level for confusion:")
print(confidence_confused)
```

```python
##test for disgusted emotion
disgusted = ['IMG_1057.jpg','IMG_1061.jpg'] #select the picture with disgusted emotion
detect = {}
confidence_disgusted = []
for name in disgusted:
    result = client.detect_faces(Image = {'S3Object':{'Bucket':'rekognition-project-files','Name':name}}, Attributes = ["ALL"])
    emotions = result['FaceDetails'][0]['Emotions']
    if emotions[0]['Type'] in detect:
        detect[emotions[0]["Type"]] += 1
    else:
        detect[emotions[0]["Type"]] = 1
    for emotions_type in emotions:
        if emotions_type["Type"] == "DISGUSTED":
            confidence_disgusted.append(emotions_type['Confidence'])

print("Computer detected:")
print(detect)
print("Confidence level for disgusted:")
print(confidence_disgusted)
```

```python
##test for calm emotion
calm = ['IMG_1043.jpg','IMG_1059.jpg','IMG_1062.jpg'] #select the picture with calm emotion
detect = {}
confidence_calm = []
for name in calm:
    result = client.detect_faces(Image = {'S3Object':{'Bucket':'rekognition-project-files','Name':name}}, Attributes = ["ALL"])
    emotions = result['FaceDetails'][0]['Emotions']
    if emotions[0]['Type'] in detect:
        detect[emotions[0]["Type"]] += 1
    else:
        detect[emotions[0]["Type"]] = 1
    for emotions_type in emotions:
        if emotions_type["Type"] == "CALM":
            confidence_calm.append(emotions_type['Confidence'])

print("Computer detected:")
print(detect)
print("Confidence level for calm:")
print(confidence_calm)
```

```python
avg_con = sum(confidence_confused)/len(confidence_confused) #average confidence level for confused emotion
avg_dis = sum(confidence_disgusted)/len(confidence_disgusted) #average confidence level for disgusted emotion
avg_cal = sum(confidence_calm)/len(confidence_calm) #average confidence level for calm emotion

Emotions = ['Confused','Disgusted','Calm']
Con_level = [avg_con, avg_dis, avg_cal]

plt.bar(Emotions, Con_level)
plt.title('Confidence level for emotions')
plt.xlabel('Emotions')
plt.ylabel('Confidence level')
plt.show()
```

From the bar chart, we can see that there are distinct differences between the confidence level for different emotions. Confused and Disgusted emotions have fewer confidence levels than Calm and Surprised. Therefore, we think that the computers are not good at detecting subtle emotions such as disgusted, but they can detect the emotions if the facial expressions have very obvious change or no change such as surprised and calm.

The data of our group member's as well as Rekognition's prediction in terms of the selected 20 images for analysis is organized in the data frame below.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("QTM 350.csv")
data
```

As mentioned above, the **Happy** to **Disgusted** columns represent the confidence level assigned by our group member to each image with the sum of them being 100. Since these images came from test banks for the emotion detection quiz, we managed to find the correct answers for the emotions represented by them and they are recorded in the **Answer** column. Given the true answer for each image, we created two columns **Our_correct_guess** and **Rekognition_correct_guess** to denote the correction rate for our group member and Rekognition respectively. For the **Our_correct_guess** column, it represents the level of confidence that our group member has assigned to the true emotion for the image, while for the **Rekognition_correct_guess**, it is the level of confidence Rekognition has given to the true emotion for the image.

Looking at the data, we can see that our group member is better than Rekognition in terms of detecting emotions from these selected images. However, we want to test quantitatively how much better our group member can detect emotions compared to Rekognition. Thus, we conducted a regression with the result for these 20 images using the **Our_correct_guess** and **Rekognition_correct_guess** columns. First, we showed the data points with a scatterplot below.

```python
human = data['Our_correct_guess']
rekognition = data["Rekognition_correct_guess"]
plt.plot(rekognition, human, 'o')
plt.xlabel('Rekognition confidence level')
plt.ylabel("Human's correct rate")
plt.show()
```

After conducting the regression of **Our_correct_guess** on **Rekognition_correct_guess**, we added the best-fit line to the graph.

```python
res = smf.ols('Our_correct_guess ~ Rekognition_correct_guess', data = data).fit()
res.params
```

```python
xs = np.array([rekognition.min(), rekognition.max()])
ys = res.params['Intercept'] + res.params["Rekognition_correct_guess"]*xs
plt.plot(rekognition, human, 'o')
plt.plot(xs,ys)
plt.xlabel('Rekognition confidence level')
plt.ylabel("Human's correct rate")
plt.show()
```

Based on the result of regression analysis, we can see that our group member makes more accurate prediction for the emotions compared to Rekognition on average (48.56% better in terms of the coefficient of the intercept). However, there seems to be no correlation between the two variables (the coefficient in front of **Rekognition_correct_guess** is very close to 0), probably due to the fact that human beings and machine have systematically different ways in terms of the detection of emotions.
