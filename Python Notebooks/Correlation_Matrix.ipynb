This short notebook shows how to create a correlation matrix.

```python
import pandas as pd
data = pd.read_csv("QTM_350.csv")
data
```

We can check the correlation between our confidence level for emotion and Rekognition's confidence level for the same emotion by first subsetting the two columns.

```python
subset = ["Our correct guess", "Rekognition's correct guess"]
matrix = data[subset]
```

Now we can make the correlation matrix.

```python
matrix.corr()
```

The correlation between each variable and itself is 1 (as expected), but the correlation between the confidence level for our guess and the confidence level of Rekognition for that same emotion is only 0.0068, implying that there is almost no correlation.
